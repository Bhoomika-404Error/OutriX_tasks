```python
import pandas as pd

# Load the uploaded file
data = pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv")

# Show first few rows
data.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(data=data, x='Churn', palette='pastel')
plt.title('Churn Count')
plt.xlabel('Did the Customer Leave?')
plt.ylabel('Number of Customers')
plt.show()
```

```python
sns.countplot(data=data, x='gender', hue='Churn', palette='Set2')
plt.title('Churn by Gender')
plt.show()
```

```python
sns.countplot(data=data, x='Contract', hue='Churn', palette='Set1')
plt.title('Churn by Contract Type')
plt.xticks(rotation=15)
plt.show()
```

```python
sns.boxplot(data=data, x='Churn', y='MonthlyCharges', palette='coolwarm')
plt.title('Monthly Charges vs Churn')
plt.show()
```

```python
import pandas as pd

data = pd.read_csv("data.csv", encoding='ISO-8859-1')

data.head()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data = data.dropna()
```

```python
data.isnull().sum()
```

```python
import matplotlib.pyplot as plt
top_products = data['Description'].value_counts().head(10)

top_products.plot(kind='bar', figsize=(10,5), color='skyblue')
plt.title("Top 10 Best-Selling Products")
plt.xlabel("Product")
plt.ylabel("Number of Times Ordered")
plt.xticks(rotation=45)
plt.show()
```

```python
top_customers = data['CustomerID'].value_counts().head(10)

top_customers.plot(kind='bar', figsize=(10,5), color='lightgreen')
plt.title("Top 10 Customers by Number of Orders")
plt.xlabel("Customer ID")
plt.ylabel("Order Count")
plt.show()
```

```python
orders_by_country = data[data['Country'] != 'United Kingdom']['Country'].value_counts().head(10)

orders_by_country.plot(kind='bar', figsize=(10,5), color='coral')
plt.title("Top 10 Countries (Excluding UK)")
plt.xlabel("Country")
plt.ylabel("Number of Orders")
plt.xticks(rotation=45)
plt.show()
```
